```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error as mse
```

```python
data = datasets.load_diabetes() #For more information please visit https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset
```

```python
data.feature_names
```

```python
data.data.shape,data.target.shape
```

```python
X = data.data
Y = data.target
```

```python
#Normalize Data
sc = StandardScaler()
X = sc.fit_transform(X)
```

# SVR Kernel
![svm kernel](svr_kernel.png)

```python
#Fit SVR with various kernel
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
svr_rbf.fit(X, Y)
svr_lin.fit(X, Y)
svr_poly.fit(X, Y)
```

```python
y_rbf = svr_rbf.predict(X)
y_lin = svr_lin.predict(X)
y_poly = svr_poly.predict(X)
```

```python
mse(Y,y_rbf),mse(Y,y_lin),mse(Y,y_poly)
```

```python
plt.figure(figsize = (14,9))
plt.plot(Y, color = 'magenta')
plt.plot(y_rbf ,color = 'green')
plt.show()
```
