```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', delimiter=',')
df.head()
```

```python
#get mean std median
df.describe()
```

```python
plt.figure(figsize = (14,9))
df_woz = df.copy()
gr = sb.countplot("bedrooms", data=df_woz, palette="bright", order = df_woz["bedrooms"].value_counts().index)
gr.set_xticklabels(gr.get_xticklabels(),rotation=90)
gr.set_title("Bedrooms Count", fontsize=20)
gr.set_xlabel("Bedrooms Number", fontsize=12)
gr.set_ylabel("Count", fontsize=12)
del df_woz #delete variable to clear memory
```

```python
plt.figure(figsize = (14,9))
df_woz = df.copy()
gr = sb.histplot(df_woz['bedrooms'])
gr.set_title("Bedrooms Distribution", fontsize=20)
gr.set_xlabel("Bedrooms Number", fontsize=12)
gr.set_ylabel("Count", fontsize=12)
del df_woz #delete variable to clear memory
```

```python
plt.figure(figsize = (14,9))
df_woz = df.copy()
gr = sb.scatterplot(data=df_woz, x="bathrooms", y="price")
gr.set_title("bathrooms vs price", fontsize=20)
gr.set_xlabel("bathrooms Number", fontsize=12)
gr.set_ylabel("price", fontsize=12)

del df_woz #delete variable to clear memory
```

```python
plt.figure(figsize = (14,9))
sb.heatmap(df.corr(), annot=True, fmt=".2f")
plt.show()
```

```python
def get_feature_name(target,n_feature):
    feature = abs(df.corr()[target]).sort_values(ascending=False)
    feature = list(feature[1:n_feature+1].keys())
    return feature
```
